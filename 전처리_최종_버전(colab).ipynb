```python
import pandas as pd
import tqdm
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
#감성말뭉치 데이터 불러오기
train_data = pd.read_excel('/content/drive/MyDrive/자연어(NLP) 프로젝트/data_folder/감성대화말뭉치(최종데이터)_Training.xlsx')
val_data = pd.read_excel('/content/drive/MyDrive/자연어(NLP) 프로젝트/data_folder/감성대화말뭉치(최종데이터)_Validation.xlsx')
all_df = pd.concat([train_data, val_data]).iloc[:, 1:]
```

```python
all_df = all_df[['사람문장1','시스템응답1','감정_대분류']]
all_df.columns = ['Q','A','label']
```

```python
# 감정 -> 0 : 중립, 1 : 나쁨 , 2: 기쁨
all_df['label'] = all_df['label'].map({'기쁨' : 2, '기쁨 ':2 ,'불안' : 1, '당황':1, '슬픔':1, '분노':1,'상처':1,'불안 ':1})
```

```python
all_df.head(2)
```

```python
PATH = '/content/drive/MyDrive/자연어(NLP) 프로젝트/data_folder/save_folder/pre_data.csv'
all_df.to_csv(PATH,index = None)
```

```python
df = pd.read_csv(PATH)
df.head(2)
```

```python
#일상대화 말뭉치 불러오기
PATH2 = '/content/drive/MyDrive/자연어(NLP) 프로젝트/data_folder/save_folder/chatbot_base.csv'
con_df = pd.read_csv(PATH2)
con_df.head(2)
```

```python
total_df = pd.concat([df,con_df])
total_df.reset_index(drop=True, inplace=True)
total_df
```

```python
#최종 전처리 데이터 저장
total_df.to_csv('ChatbotData.csv', index = None)
```
